# 04. Aggregating Data with agg()

## Performing a single aggregation

```python
#### Using agg() to calculate the mean of 'order_number' column grouped by 'department_id'
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#### Using mean() to calculate the mean of 'order_number' column grouped by 'department_id'
df.groupby('department_id')['order_number'].mean()
```

```python
#### Calculating sum of averages from 'order_number' column (Relevant for Q2 of Task for this Exercise)
result = df.groupby('department_id')['order_number'].mean().sum()
```

```python
print(result)
```

## Performing Multiple Aggregations

```python
#### Using agg() to calculate the mean, min & max of 'order_number' column grouped by 'department_id'
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# 05. Aggregating Data with transform()

## Creating a 'max_order' column in order to the end of creating flags for loyalty customers

```python
#### Creating 'max_order' column in 'ords_prods_merge' df
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

# 06. Deriving Columns with loc()

## Create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
