### Create Conditions: If-Statements with User-Defined Functions

```python
#### Using head function to get the first 1,000,000 rows
df_subset = df.head(1000000)
```

```python
print(type(df_subset))
```

```python
#### Define a function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif 5 < row['prices'] <= 15:
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else:
        return 'Not enough data'
```

```python
#### Apply the function
df_subset['price_range'] = df_subset.apply(price_label, axis=1)
```

```python
df_subset['price_range'].value_counts(dropna=False)
```

```python
#### Checking most expensive product in df_subset
df_subset['prices'].max()
```

### Create Conditions: If-Statements with the loc() function

```python
df_subset.loc[df_subset['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_subset.loc[(df_subset['prices'] <= 15) & (df_subset['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_subset.loc[df_subset['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### Create Conditions: If-Statements with For-Loops

```python
#### Print frequency of 'orders_day_of_week' column in order to determine day on which most orders occurs
df['orders_day_of_week'].value_counts(dropna = False)
```

```python
#### Creating new column,'busiest day', that will contain one of three different values: 'Busiest day', 'Least busy', and 'Regularly busy'.

result = []

for value in df["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#### Adding the values of 'result' into a new column, 'busiest day' in the 'df' dataframe
df['busiest_day'] = result
```

```python
df['busiest_day'].value_counts(dropna = False)
```

## S1. Update 'busiest day' column to 'busiest days'

```python
#### Perform update of 'busiest_day' column to busiest_days" from "orders_day_of_week" to identify the two busiest days (0, 1), the two slowest days (4, 3), and all other days (regularly busy).

result_2 = []

for value in df["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

```python
result_2
```

```python
#### Create new column 'busiest_days' in df dataframe
df['busiest_days'] = result_2
```

```python
df['busiest_days'].value_counts(dropna = False)
```

## S3. Create 'busiest_period_of_day' column

```python
#### Check value counts in "time_of_day" column.

df['time_of_day'].value_counts()
```

#### Comments on 'value_counts' for 'time_of_day': Busiest hours of the day will be identified by number of orders as reflected in the 'value_counts'. There will be group of values into the following three categories - (1) Most orders (2) Average Orders (3) Fewest Orders.

```python
#### Creating If-Statement with For-Loop in order to group value_counts for time_of_day into Most orders, Average orders, and Fewest orders.

result_3 = []

for value in df['time_of_day']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append('Most orders')
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append('Fewest orders')
  else:
    result_3.append('Average orders')
```

```python
result_3
```

```python
#### Creating 'busiest_period_of_day' column in df dataframe.

df['busiest_period_of_day'] = result_3
```

```python
#### Print value_counts in 'busiest_period_of_day' column.

df['busiest_period_of_day'].value_counts(dropna = False)
```
