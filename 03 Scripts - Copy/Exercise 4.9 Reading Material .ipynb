# 05. Creating Line Charts & Sampling Data

## Splitting 'ords_prods_merge' dataframe into two randomized samples at a 70/30 ratio

```python
#### Create a list holding True/False values to the test np.random.rand() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
#### Store 70% of the sample into the dataframe big
big = ords_prods_merge[dev]
```

```python
#### Store 30% of the sample into the dataframe small
small = ords_prods_merge[~dev]
```

## Check to see if two new sample subsets (i.e. big & small) equal original dataframe

```python
#### Check for exact number of rows between new sample subsets and 'ords_prods_merge' df using len()
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

## Speeding up processing time by reducing sample to columns necessary for chart 

```python
#### Creating a new dataframe, df_2, that contains only the 'prices' and 'orders_day_of_week' columns from small subset
df_2 = small[['orders_day_of_week','prices']]
```

## Create line chart

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

## Export the line chart

```python
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'linechart_orders_dow.png'))
```
